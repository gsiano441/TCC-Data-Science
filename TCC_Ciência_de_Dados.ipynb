```python
X = data.drop('target', axis=1)
X
```

```python
y=data.target
y
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
#Utilizando o RandomUnderSampler para reduzir a quantidade de dados, selecionando o parâmetro de uma proporção benigno/maligno=0,7
rus = RandomUnderSampler(sampling_strategy=0.7, random_state=42)
X_res, y_res = rus.fit_resample(X, y)
```

```python
#Criando os datasets de treino e teste utilizando o stratify para manter a proporção entre as classes em todos os datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.25, random_state=42, stratify=y_res)
```

```python
#Separando as classes para salvar as imagens em pastas rotuladas com as classes
X_train_benign = X_train[X_train['benign_malignant']=='benign']
X_train_malignant = X_train[X_train['benign_malignant']=='malignant']

X_test_benign = X_test[X_test['benign_malignant']=='benign']
X_test_malignant = X_test[X_test['benign_malignant']=='malignant']
```

```python
#Criando listas com os nomes das imagens para salvar por meio de prompt de comando
lista_train_benign = X_train_benign['image_name'].tolist()
lista_train_malignant = X_train_malignant['image_name'].tolist()
lista_test_benign = X_test_benign['image_name'].tolist()
lista_test_malignant = X_test_malignant['image_name'].tolist()

print(len(lista_train_benign))
print(len(lista_test_benign))
print(len(lista_train_malignant))
print(len(lista_test_malignant))
```

```python
#Transformando as listas com os nomes em caminhos para a pasta de destino
caminho_base = "C:\\Users\\Gabriela Siano\\Downloads\\jpeg\\train\\"
tipo = ".jpg"

#Criando as listas com os nomes das imagens
enderecos_train_benign = [caminho_base + nome + tipo for nome in lista_train_benign]
enderecos_train_malignant = [caminho_base + nome + tipo for nome in lista_train_malignant]
enderecos_test_benign = [caminho_base + nome + tipo for nome in lista_test_benign]
enderecos_test_malignant = [caminho_base + nome + tipo for nome in lista_test_malignant]

#Exibindo a lista
print(enderecos_train_benign)
print(enderecos_train_malignant)
print(enderecos_test_benign)
print(enderecos_test_malignant)
```

Criação da Rede Neural

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
dir_treino = '/content/drive/MyDrive/TCC Ciência de Dados/TCC_classes/train'
dir_teste = '/content/drive/MyDrive/TCC Ciência de Dados/TCC_classes/test'
```

```python
data_generator = ImageDataGenerator(rescale=1/255.0)
```

```python
gerador_treino = data_generator.flow_from_directory(dir_treino, batch_size=32, target_size=(224, 224), class_mode='sparse')
```

```python
gerador_teste = data_generator.flow_from_directory(dir_teste, batch_size=32, target_size=(224, 224), class_mode='sparse')
```

```python
model = tf.keras.applications.Xception(input_shape=(224,224,3), include_top=False)
```

```python
average_pooling = tf.keras.layers.GlobalAveragePooling2D()(model.output)
```

```python
hidden = tf.keras.layers.Dense(2048, activation='relu')(average_pooling)
```

```python
pred = tf.keras.layers.Dense(1, activation='sigmoid')(hidden)
```

```python
model_final = tf.keras.models.Model(inputs = model.input, outputs=pred)
```

```python
#sumário do modelo
model_final.summary()
```

```python
#Compilando o modelo
model_final.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model_final.fit(gerador_treino, epochs=100)
```

```python
#Avaliando o modelo
model_final.evaluate(gerador_teste)
```
